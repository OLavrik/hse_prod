## Process

```python
train_df1 = pd.read_csv('/Users/Olga.Lavrichenko/Documents/ВШЭ/prod/1/train01.csv', sep=';')
train_df2 = pd.read_csv('/Users/Olga.Lavrichenko/Documents/ВШЭ/prod/1/train02.csv', sep=';')
test_df = pd.read_csv('/Users/Olga.Lavrichenko/Documents/ВШЭ/prod/1/test.csv', sep=';')

train_df1['split'] = 'train'
train_df2['split'] = 'train'
test_df['split'] = 'test'
train_df2 = train_df2[~(train_df2.id.isin(train_df1.id))].copy()
train_df1 = train_df1.reset_index(drop=True)
train_df2 = train_df2.reset_index(drop=True)
test_df = test_df.reset_index(drop=True)
train_df1.loc[~train_df1['x2'].isnull(), 'x2'] = train_df1.loc[~train_df1['x2'].isnull(), 'x2'].apply(lambda x: datetime.strptime(x, '%d.%m.%Y'))
train_df2.loc[~train_df2['x2'].isnull(), 'x2'] = train_df2.loc[~train_df2['x2'].isnull(), 'x2'].apply(lambda x: datetime.strptime(x, '%d.%m.%Y'))
test_df.loc[~test_df['x2'].isnull(), 'x2'] = test_df.loc[~test_df['x2'].isnull(), 'x2'].apply(lambda x: datetime.strptime(x, '%Y-%m-%d %H:%M:%S'))




```

```python
base_ = {
    'min': partial(np.nanmin, axis=1),
    'max': partial(np.nanmax, axis=1),
    'mean': partial(np.nanmean, axis=1),
    'median': partial(np.nanmedian, axis=1),
    'var': partial(np.nanvar, axis=1),
    'skew': lambda x: ((x - np.nanmean(x, axis=1, keepdims=True)) ** 3).mean(1) / np.nanstd(x, axis=1) ** 3,
    'kurt': lambda x: ((x - np.nanmean(x, axis=1, keepdims=True)) ** 4).mean(1) / np.nanstd(x, axis=1) ** 4,
    'p_25': partial(np.nanpercentile, q=25, axis=1),
    'p_75': partial(np.nanpercentile, q=75, axis=1),
    'p_nans': lambda x: np.isnan(x).mean(axis=1)
}

def proc(df):
    for col in ['x8', 'x9']:
        df_col = np.array(df[col].tolist())
        res = np.hstack([aggr(df_col).reshape(-1, 1) for aggr in base_.values()])
        res[np.isnan(res)] = 0.
        df = pd.concat([df, pd.DataFrame(res, columns=[col + '_' + k for k in base_.keys()])], axis=1)
    return df



train_df1 = proc(train_df1)
train_df2 = proc(train_df2)
test_df = proc(test_df)
```

```python





df = pd.concat([train_df1, train_df2, test_df])
del train_df1, train_df2, test_df
df.drop(['x8', 'x9'], axis=1, inplace=True)
nan_ratio = {}
for col in tqdm(df.columns):
    if col != 'y':
        nan_ratio[col] = df[col].explode().isnull().mean()

for k, v in nan_ratio.items():
    if v > 0.7:
        df.drop(k, axis=1, inplace=True)


df = df.reset_index(drop=True)

for col in ['x13', 'x14', 'x15', 'x16', 'x17', 'x18', 'x19', 'x20', 'x21', 'x22', 'x26', 'x34', 'x35', 'x36', 'x38', 'x39']:
    item = df.loc[0, col]
    len_ = len(item)
    values = np.array(df[col].tolist())
    values = np.hstack([values, np.isnan(values).mean(1, keepdims=True)])
    values[np.isnan(values)] = 0.
    col_names = [f'{col}_{i}' for i in range(1, len_ + 1)] + [f'{col}_p_nans']
    df[col_names] = values
    df.drop(col, axis=1, inplace=True)

df.loc[df['x2'].isnull(), 'x2'] = datetime(1, 1, 1, 0, 0, 0)
df.loc[df['x5'].isnull(), 'x5'] = -1

df['x2_month'] = df['x2'].apply(lambda x: x.month)
df['x2_day'] = df['x2'].apply(lambda x: x.day)



for col in ['x1', 'x3', 'x5', 'x6']:
    df[col] = LabelEncoder().fit_transform(df[col].values)
for col in df.columns:
    if df[col].dtype == np.int64:
        df[col] = df[col].astype(np.int32)
    if df[col].dtype == np.float64:
        df[col] = df[col].astype(np.float32)

df.drop(['id', 'x2'], axis=1, inplace=True)

train = df.loc[df['split'] == 'train'].copy()
test_df = df.loc[df['split'] == 'test'].copy()

train.drop('split', axis=1, inplace=True)
test_df.drop('split', axis=1, inplace=True)


X_train = train.iloc[:, 1:].values
y_train = train['y'].values


X_test = test_df.iloc[:, 1:].values
```

```python
def write_res(y_pred):
    sub = pd.DataFrame(np.hstack([np.arange(len(y_pred)).reshape(-1, 1), y_pred[:, 1, None]]), columns=['id', 'result'])
    sub['id'] = sub['id'].astype(int)
    sub = sub.set_index('id')
    sub.to_csv('submission_catboost.csv',  index='id')
```
